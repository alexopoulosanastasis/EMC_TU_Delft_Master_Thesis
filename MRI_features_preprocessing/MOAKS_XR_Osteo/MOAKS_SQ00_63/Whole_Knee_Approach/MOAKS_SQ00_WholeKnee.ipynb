# NOW WE DELETE ALL THE COMMENTS OF THE INITIAL DATAFRAME AND CONVERT THE VALUES TO NUMBERS

```python
# Firstly we replace all the comments
moaks_without_commments = moaks_sq00.replace(r'(\:).*$',r'\1',regex=True)
# Secondly we replace all the '.:' and the ':' bullets with space and remove 'R' from Reader Column
moaks_without_commments_bullets = moaks_without_commments.replace({'\.\:':'',':':''},regex=True)
# Thirdly we remove the Comments and Technical Considerations Columns from the whole Dataset
moaks_without_commments_bullets = moaks_without_commments_bullets.drop(columns=['V00MCMNTS','V00MTCMNTS'])
print(moaks_without_commments_bullets.READPRJ.value_counts())
```

```python
# WE KEEP ONLY THE SUBJECTS THAT ARE PRESENT IN PROJECT 63F AND NOT IN PROJECT 22

moaks_without_commments_bullets_prj22 = moaks_without_commments_bullets.loc[moaks_without_commments_bullets['READPRJ']=='22']
print(moaks_without_commments_bullets_prj22.shape[0])

moaks_without_commments_bullets_prj63F = moaks_without_commments_bullets.loc[moaks_without_commments_bullets['READPRJ']=='63F']
print('Number of subjects with OA at baseline prj63: {}'.format(moaks_without_commments_bullets_prj63F.shape[0]))
moaks_without_commments_bullets_prj63F_unique_ID = pd.merge(moaks_without_commments_bullets_prj63F,moaks_without_commments_bullets_prj22,on=['ID','SIDE'],how='outer',indicator=True)
moaks_without_commments_bullets_prj63F_unique_ID = moaks_without_commments_bullets_prj63F_unique_ID.loc[(moaks_without_commments_bullets_prj63F_unique_ID._merge=='left_only'),['ID','SIDE']]
moaks_without_commments_bullets_prj63F_unique_info = moaks_without_commments_bullets_prj63F.merge(moaks_without_commments_bullets_prj63F_unique_ID,on=['ID','SIDE'],how='inner')
moaks_without_commments_bullets_prj63F_unique_info = moaks_without_commments_bullets_prj63F_unique_info.drop_duplicates(subset=['ID','SIDE'])
print('Number of subjects with OA at baseline unique to prj63: {}'.format(moaks_without_commments_bullets_prj63F_unique_info.shape[0]))

moaks_without_commments_bullets_prj63E = moaks_without_commments_bullets.loc[moaks_without_commments_bullets['READPRJ']=='63E']
print('Number of subjects without OA at baseline prj63: {}'.format(moaks_without_commments_bullets_prj63E.shape[0]))
moaks_without_commments_bullets_prj63E_unique_ID = pd.merge(moaks_without_commments_bullets_prj63E,moaks_without_commments_bullets_prj22,on=['ID','SIDE'],how='outer',indicator=True)
moaks_without_commments_bullets_prj63E_unique_ID = moaks_without_commments_bullets_prj63E_unique_ID.loc[(moaks_without_commments_bullets_prj63E_unique_ID._merge=='left_only'),['ID','SIDE']]
moaks_without_commments_bullets_prj63E_unique_info = moaks_without_commments_bullets_prj63E.merge(moaks_without_commments_bullets_prj63E_unique_ID,on=['ID','SIDE'],how='inner')
moaks_without_commments_bullets_prj63E_unique_info = moaks_without_commments_bullets_prj63E_unique_info.drop_duplicates(subset=['ID','SIDE'])
print('Number of subjects without OA at baseline unique to prj63: {}'.format(moaks_without_commments_bullets_prj63E_unique_info.shape[0]))
```
