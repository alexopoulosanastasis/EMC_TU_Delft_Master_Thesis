```python
common_fnih_kxr_00 = pd.merge(fnih_sq00,KL_prj15_osteophytes,on=['ID','SIDE'],how='inner')
osteo_fnih_labels = ['ID','SIDE','V00MOSFMA','V00MOSFLA','V00MOSFMP','V00MOSFLP','V00MOSFMC','V00MOSFLC','V00MOSTM','V00MOSTL',
                    'V00XROSFL','V00XROSFM','V00XROSTL','V00XROSTM','V00XRKL']

common_fnih_kxr_00 = common_fnih_kxr_00.loc[:,osteo_fnih_labels]
common_fnih_kxr_00.fillna(0.0,inplace=True)
common_fnih_kxr_00
```

```python
common_fnih_kxr_00_NAN = common_fnih_kxr_00.loc[common_fnih_kxr_00.V00XROSFL.isna()]
print('Number of NaN values in combined Xray & MOAKS df is: {}'.format(common_fnih_kxr_00_NAN.shape[0]))
```
