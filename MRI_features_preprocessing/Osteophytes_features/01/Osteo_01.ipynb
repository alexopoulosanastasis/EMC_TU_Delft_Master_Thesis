```python
fnih_sq01 = pd.read_csv('/home/anastasis/EMC_Thesis/My_codes/MOAKS/MOAKS_FNIH_SQ01/Whole_Knee_Approach/moaks_fnih_sq01_initial_numeric_df.csv',index_col=0)
fnih_sq01
```

```python
common_fnih_kxr_01 = pd.merge(fnih_sq01,KL_prj15_osteophytes,on=['ID','SIDE'],how='inner')
osteo_fnih_labels = ['ID','SIDE','V01MOSFMA','V01MOSFLA','V01MOSFMP','V01MOSFLP','V01MOSFMC','V01MOSFLC','V01MOSTM','V01MOSTL',
                    'V01XROSFL','V01XROSFM','V01XROSTL','V01XROSTM','V01XRKL']

common_fnih_kxr_01 = common_fnih_kxr_01.loc[:,osteo_fnih_labels]
common_fnih_kxr_01.fillna(0.0,inplace=True)
common_fnih_kxr_01
```

```python
common_fnih_kxr_01_NAN = common_fnih_kxr_01.loc[common_fnih_kxr_01.V01XROSFL.isna()]
print('Number of NaN values in combined Xray & MOAKS df is: {}'.format(common_fnih_kxr_01_NAN.shape[0]))
```
